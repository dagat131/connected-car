```python
import pandas as pd
import numpy as np
# read_csv 함수로 데이터를 Dataframe 형태로 불러옵니다.
file_path = 'C:/python/here/data/2019.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv'
chipo = pd.read_csv(file_path)
chipo
```

```python
print(chipo.shape)
print("------------------------------------")
print(chipo.info())
```

```python
for i in range(1,5):
    chipo[chipo.columns[i]] =chipo[chipo.columns[i]].map(lambda x: x.replace(",",""))
    chipo[chipo.columns[i]] = chipo[chipo.columns[i]].astype(int)
```

```python
print(chipo.info())
```

```python
for i in range(7,28):
    chipo[chipo.columns[i]] =chipo[chipo.columns[i]].map(lambda x: x.replace(",",""))
    chipo[chipo.columns[i]] = chipo[chipo.columns[i]].astype(int)
```

```python
print(chipo.info())
```

```python
chipo["행정구역"] =chipo["행정구역"].map(lambda x: x.replace("세",""))
chipo["행정구역"] =chipo["행정구역"].map(lambda x: x.replace("이상",""))

chipo["행정구역"] = chipo["행정구역"].astype(int)
```

```python
print(chipo.info())
```

```python
chipo1 = np.array(chipo["총계"])
freq, _ = np.histogram(chipo1, bins=10, range=(0, 50000))
freq
```

```python
freq_class = [f'{i}~{i+5000}' for i in range(0, 50000,5000)]
# freq_class를 인덱스로 DataFrame을 작성
freq_dist_df = pd.DataFrame({'frequency':freq},
                            index=pd.Index(freq_class,name='인구'))
freq_dist_df
```

```python
import pandas as pd
import numpy as np
# read_csv 함수로 데이터를 Dataframe 형태로 불러옵니다.
D1806 = pd.read_csv('2018.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv',thousands=',')
D1812 = pd.read_csv('2018.12.31.기준 대구광역시 주민등록인구 현황_UTF8.csv',thousands=',')
D1906 = pd.read_csv('2019.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv',thousands=',')

print(D1806)
print(D1812)
print(D1906)
```

```python
print("------------------------------------")
print(D1806.info())
print("------------------------------------")
print(D1812.info())
print("------------------------------------")
print(D1906.info())
```

```python
D1806_1 = np.array(D1806)
D1806_11 = np.delete(D1806_1,(0), axis=1)
D1812_2 = np.array(D1812)
D1812_22 = np.delete(D1812_2,(0), axis=1)
D1906_3 = np.array(D1906)
D1906_33 = np.delete(D1906_3,(0), axis=1)

a=np.subtract(D1806_11,D1812_22)
a.shape
b=np.ones((101,27),dtype='<U11')

print(type(b[1][1]))
print(type(a[1][1]))
```

```python

for i in range(0,101):
    for j in range(0,27):
        if a[i][j]>0:
            b[i][j]=str(a[i][j])+"증가"
        elif a[i][j]<0:
            b[i][j]=str(a[i][j]*(-1))+"감소"
        else :
            b[i][j]="인구유지"
print(b)
```

```python
df =pd.DataFrame(a,columns=D1806.columns[1:])
df
```

```python
df1 =pd.DataFrame(b,columns=D1806.columns[1:])
df1
```

```python
import matplotlib.pyplot as plt
x_set=np.array(df.index)
prob=np.array(df["총계"])
prob1=np.zeros(101,)
prob2=np.zeros(101,)
for i in range(0,101):
    if prob[i]>=0:
        prob1[i]=prob[i]
    else:
        prob1[i]=0

for i in range(0,101):
    if prob[i]<0:
        prob2[i]=prob[i]
    else:
        prob2[i]=0


fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111)
ax.bar(x_set, prob1)
ax.bar(x_set, prob2,color='r')
ax.set_xlabel('age',size=20)
ax.set_ylabel('p&m',size=20)

plt.show()
```
